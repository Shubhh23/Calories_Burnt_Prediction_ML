Converting the text data to numerical values

```python
calories_data.replace({"Gender":{'male':0 , 'female':1}}, inplace=True)
```

```python
correlation = calories_data.corr()
```

```python
# constructing a heatmap to understand the correlation

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True ,annot_kws={'size':8}, cmap='Blues')
```

Separating features and Target

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size=0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

Feature Importance

```python
importance = model.feature_importances_
print(importance)
```

```python
import matplotlib.pyplot as plt

features = X.columns

plt.bar(features, importance)
plt.xticks(rotation=45)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.title("Feature Importance")
plt.show()
```

Evaluation

Prediction on Test Data

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

Actual vs Predicted Graph

```python
plt.figure(figsize=(6,6))

plt.scatter(Y_test, test_data_prediction)

plt.xlabel("Actual Calories")

plt.ylabel("Predicted Calories")

plt.title("Actual vs Predicted")

plt.show()
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean Absolute Error: ", mae)
```

R² Score


```python
from sklearn.metrics import r2_score

score = r2_score(Y_test, test_data_prediction)
print(score)
```

```python
import numpy as np

age = float(input("Enter Age: "))
gender = int(input("Gender (Male=1, Female=0): "))
height = float(input("Height (cm): "))
weight = float(input("Weight (kg): "))
duration = float(input("Exercise Duration (minutes): "))
heart_rate = float(input("Heart Rate: "))
body_temp = float(input("Body Temperature: "))

input_data = (age, gender, height, weight, duration, heart_rate, body_temp)

input_array = np.asarray(input_data).reshape(1, -1)

prediction = model.predict(input_array)

print("\nEstimated Calories Burned:", round(prediction[0], 2), "kcal")
```

```python
#Q1. Why do we use np.asarray()?
#Answer: It converts the input into a NumPy array because Scikit-learn models work efficiently with NumPy arrays.

#Q2. Why do we use reshape(1, -1)?
#Answer: The model expects input in a 2D format, where each row represents one sample and each column represents a feature. 
#Since we are predicting for one person, we reshape the input to one row with seven features.

#Q3. Why do we use prediction[0]?
#Answer:model.predict() returns a NumPy array, even for a single prediction. 
#We use prediction[0] to extract the predicted calorie value from that array.

#⭐ Interview Tip
#When an interviewer asks, "How does your predictive system work?", you can answer:
#"After training the Random Forest model, I built a predictive system where the user enters seven features:
#age, gender, height, weight, exercise duration,heart rate, and body temperature. 
#I convert these values into a NumPy array, reshape them into the required 2D format, and pass them to the trained model using model.predict(). 
#The model then estimates the number of calories burned, which is displayed to the user."
```
